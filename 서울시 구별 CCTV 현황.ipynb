```python
import pandas as pd
```

```python
# 서울시 자치구 연도별 cctv 설비 현황 데이터 불러오기

CCTV_Seoul = pd.read_excel('../data/01. CCTV_in_Seoul.xlsx', header=2)

CCTV_Seoul.head()
```

```python
CCTV_Seoul.drop([0], inplace=True)
CCTV_Seoul.drop(CCTV_Seoul.columns[[0,1,-1]], axis=1, inplace=True)

CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0] : '구별',
                           CCTV_Seoul.columns[1] : 'CCTV 총계'}, inplace=True)



CCTV_Seoul.head()
```

```python
CCTV_Seoul.insert(2, '2020년도 이전', CCTV_Seoul.iloc[:,2:8].sum(axis=1))
CCTV_Seoul.drop(CCTV_Seoul.columns[3:9], axis=1, inplace=True)
CCTV_Seoul.head()
```

```python
# 서울시 자치구 등록인구 데이터 불러기기

pop_Seoul = pd.read_csv('../data/01. population_in_Seoul.csv',
                       header = 2,
                       usecols = [1, 2, 4, 7, 10])

pop_Seoul.head()
```

```python
pop_Seoul = pop_Seoul[pop_Seoul.iloc[:, 1] == '소계']
pop_Seoul.drop(pop_Seoul.columns[1], axis=1, inplace=True)
pop_Seoul.drop([0], inplace=True)
pop_Seoul.reset_index(drop=True, inplace=True)

pop_Seoul.rename(columns={pop_Seoul.columns[0] : '구별',
                          pop_Seoul.columns[1] : '인구수',
                          pop_Seoul.columns[2] : '한국인',
                          pop_Seoul.columns[3] : '외국인'}, inplace=True)

pop_Seoul.head()
```

```python
pop_Seoul.rename(columns={pop_Seoul.columns[0] : '구별',
                          pop_Seoul.columns[1] : '인구수',
                          pop_Seoul.columns[2] : '한국인',
                          pop_Seoul.columns[3] : '외국인'}, inplace=True)

pop_Seoul.head()
```

```python
# 서울시 외국인비율 가장 높은 구 5개

pop_Seoul['외국인비율'] = pop_Seoul['외국인'] / pop_Seoul['인구수'] * 100
pop_Seoul.sort_values(by='외국인비율', ascending=False).head()
```

```python
# ccvt 데이터 및 인구 현황 데이터 합기기

data_result = pd.merge(CCTV_Seoul, pop_Seoul, on='구별')

data_result.head()
```

```python
data_result.drop(data_result.columns[2:6], axis=1, inplace=True)
data_result.set_index('구별', inplace=True)

data_result.head()
```

```python
# 각 구별 cctv 총계 그프프

data_result['CCTV 총계'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.xlabel('CCTV 총계')
plt.show()
```

```python
# 각 구별 인구수 대비 cctv 비율 그래프

data_result['CCTV 비율'] = data_result['CCTV 총계'] / data_result['인구수'] * 100
data_result['CCTV 비율'].sort_values().plot(kind='barh', grid=True, figsize=(10, 10))
plt.xlabel('CCTV 비율')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['CCTV 총계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV 총계')
plt.grid()
plt.show()
```

```python
# 인구수 및 cctv 데이터의 최적 1차다항식(선형 회귀) 구하기

fp1 = np.polyfit(data_result['인구수'], data_result['CCTV 총계'], 1)
fp1
```

```python
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['CCTV 총계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV 총계')
plt.grid()
plt.show()
```

```python
# cctv 실제값과 선형회귀값의 오차 구하기

data_result['오차'] = np.abs(data_result['CCTV 총계'] - f1(data_result['인구수']))

df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'], data_result['CCTV 총계'],
            c=data_result['오차'], s=50, cmap='seismic')
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

# 상위 10개 데이터 포인트에 인덱스값을 텍스트로 표시 (오른쪽 아래 좌표에 표시)
for n in range(10):
    plt.text(df_sort.iloc[n]['인구수']*1.02, df_sort.iloc[n]['CCTV 총계']*0.98,
             df_sort.index[n], fontsize=15)
    
plt.xlabel('인구수')
plt.ylabel('CCTV 총계')

plt.colorbar()
plt.grid()
plt.show()
```
