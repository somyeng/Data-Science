```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
# "tips" : seaborn     , , , ,   
tips = sns.load_dataset('tips')
tips.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
```

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip', data=tips)
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', hue='smoker', data=tips, palette='Set1')
plt.show()
```

```python
#     
flights = sns.load_dataset('flights')
flights.head()
```

```python
flights_pv = flights.pivot_table(values=['passengers'], index=['month'], columns=['year'])
flights_pv.columns = flights_pv.columns.droplevel(0)
flights_pv.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights_pv, annot=True, fmt='f')
plt.show()
```
