```python
from bs4 import BeautifulSoup
```

```python
page = open('../data/999. BeautifulSoup_test_first.html','r').read()
soup = BeautifulSoup(page, 'html.parser') # soup 변수 = 전체 html 코드
print(soup.prettify()) # prettify() : 페이지 내용 전체 출력
```

```python
list(soup.children) # soup의 자식 태그 리트트
```

```python
html = list(soup.children)[2] # <html> 태그 내용 : soup의 자식 태그
html
```

```python
list(html.children) # <html>의 자식 태그 리스트
```

```python
body = list(html.children)[3] # <body> 태그 내용 : <html>의 자식 태그
body
```

```python
soup.body # <body> 태그 내용에 바로 접근하기
```

```python
list(body.children) # <body>의 자식 태그 리스트
```

```python
for each_tag in soup.find_all('p'):
    print(each_tag.get_text()) # <p> 전체 태그의 text값 출력
```

```python
links = soup.find_all('a') # <a> 전체 태그 내용
links
```

```python
for each in links :
    href = each['href']
    text = each.string
    print(text + ' -> ' + href)  # <a> 전체 태그의 'href' 및 text 출력
```
